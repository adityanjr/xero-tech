## Flipping

```python
import os
import io

def flipHorizontal():
    Annotation_to_flip = input("Enter the path of the directory of annotations")  # directory of the annotations
    write_flip_annotations = input("Enter the path of the directory of flip_annotations") # directory to write flipped annotations
    dir_list = os.listdir(Annotation_to_flip)
    #dir_list.remove("classes.txt")  # uncomment accordingly
    print(dir_list)
    for file_name in dir_list:    
        file = open(os.path.join(Annotation_to_flip,file_name),'r')
        try: 
            for line in file.readlines():
                entity_list = line.split()
                print(entity_list)
                x = (2*(0.5 - float(entity_list[1])) + float(entity_list[1]))
                x = round(x, 6)
                entity_list[1] = str(x) 
                print(x)   
                print(entity_list)
                file_write = open(os.path.join(write_flip_annotations,file_name),'a')
                for items in entity_list:
                    file_write.write(items)
                    file_write.write(' ')
        except:
            print("Cannot access")
    file.close()

flipHorizontal()
    
    
```
